```python
D['Ram']=D['Ram'].str.replace('GB','')
```

```python
D['Weight']=D['Weight'].str.replace('kg','')
```

```python
new = D["ScreenResolution"].str.split("x", n = 1, expand = True)
```

```python
new
```

```python
D['X_res']=new[0]
D['Y_res']=new[1]
```

```python
D["X_res"]=(D['X_res'].str.replace(',','').str.findall(r'(\d+\d+\d+\d+)').apply(lambda x: pd.Series(x).astype(int)).mean(1))
```

```python
D["X_res"]=pd.to_numeric(D["X_res"])
D["Y_res"]=pd.to_numeric(D["Y_res"])
```

```python
D["PPI"]=(((D["X_res"]**2+D["Y_res"]**2)**(1/2))/D["Inches"]).astype(int)
```

```python
D
```

```python
D["ScreenResolution"]=(D["X_res"]*D["Y_res"]).astype(float)
```
