```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 15,8
p = sns.countplot(DF['Company'], palette = 'pastel')
plt.title('Brand', fontsize = 20)
plt.xlabel('Brands')
plt.ylabel('Count')
plt.show()
```
